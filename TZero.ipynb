```python
#################################################
# Python script for T-Zero fight-or-flight game #
# Author: Rolando Gonzales Martinez             #
# Version: Beta (October 25, 2024)              #
#################################################
import random
import numpy as np
import time

# Hyperbolic tangent neural network for T-Zero attack probability
def attack_probability(runs, attacks):
    input_vector = np.array([runs, attacks])
    weights = np.array([0.6, 0.4])  # Weights to emphasize running and attacking
    probability = np.tanh(np.dot(weights, input_vector))
    return probability

# Game variables
human_health = 100
terminator_health = 100
run_count = 0
attack_count = 0
start_time = time.time()

# ASCII representations
def display_status():
    # Convert health values to integers for proper display
    human_health_int = int(human_health)
    terminator_health_int = int(terminator_health)

    print("\nHuman Player:  ðŸ˜Š ", f"[{'#' * (human_health_int // 10)}{' ' * (10 - human_health_int // 10)}] {human_health_int}%")
    print("T-Zero:        ðŸ¤– ", f"[{'#' * (terminator_health_int // 10)}{' ' * (10 - terminator_health_int // 10)}] {terminator_health_int}%\n")

def game_round():
    global human_health, terminator_health, run_count, attack_count

    # Display choices for the player
    choice = input("Choose your action:\n1. Fight\n2. Run\n> ")

    if choice == '2':  # Run
        print("You chose to run away!")
        run_count += 1
        attack_probability_value = attack_probability(run_count, attack_count)
        if random.random() < attack_probability_value:
            damage = random.randint(5, 15)
            print(f"T-Zero attacks while you flee, causing {damage}% damage!")
            human_health -= damage
    elif choice == '1':  # Fight
        print("You chose to fight!")
        attack_count += 1
        weapon = random.choices(['pistol', 'shotgun', 'grenade'], [0.5, 0.4, 0.1])[0]

        if weapon == 'pistol':
            damage = random.uniform(5, 10)
            print(f"You got a pistol! You caused {damage:.2f}% damage to T-Zero.")
            terminator_health -= damage
        elif weapon == 'shotgun':
            damage = random.uniform(25, 40)
            print(f"You got a shotgun! You caused {damage:.2f}% damage to T-Zero.")
            terminator_health -= damage
            counter_damage = random.uniform(30, 40)
            print(f"T-Zero fights back, causing {counter_damage:.2f}% damage to you!")
            human_health -= counter_damage
        elif weapon == 'grenade':
            damage = 30
            print(f"You got a grenade! You caused {damage}% damage to T-Zero.")
            terminator_health -= damage

    # Display current health status
    display_status()

# Game introduction
print("""
================================================================
    ***      TERMINATOR ZERO: FIGHT-OR-FLIGHT    ****
   Run or fight. Try to survive as much time as possible
â €â €â €â €â €â €â €â €â €â €â €â¢ â¡žâ¢â£´â£¶â£¿â£»â¡·â €â£‡â£€â£€â¡€â €â¢°â¡‡â €â ˆâ¢³â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢€
â €â €â €â €â €â €â €â €â €â €â €â¢¸â â ¿â£·â ¿â¡Ÿâ ‹â €â €â €â €â €â €â ˆâ ‰â â£¦â£¤â£¸â¡†â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢€â£¿â €â €â ˜â ‰â â¡„â €â €â €â €â €â €â¡„â €â €â¢¸â£¿â£¾â£¿â¡€â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â¢¿â €â €â €â €â €â¢·â €â €â €â €â €â¢€â¡‡â €â €â ˆâ ¹â¢¿â ºâ¡‡â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â£¸â¡‡â €â €â €â €â¢¸â£´â¢³â €â¢ â¢»â£¼â €â €â €â €â €â£¼â¢¸â¡‡â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â£¿â ›â ¤â¢€â£€â €â¡€â ¿â£¼â£„â£¸â£¼â Ÿâ£€â£€â¡€â ¤â –â »â£¿â¡‡â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â¡â£‡â¡€â €â ˆâ ‰â ™â “â ¬â ­â ­â ”â šâ ‰â â €â£€â£€â£ â£¿â â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â£«â£¿â£­â£½â£·â£¦â£¤â£€â €â €â €â£€â£¤â£´â£¶â£¿â£·â£¶â£·â£½â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢°
â €â €â €â €â €â €â €â €â €â €â¢¸â »â¢¸â£¿â£¿â¡ºâ¡½â¢»â£¿â£¿â¡¤â£¼â£¿â£¿â£»â¡¾â£»â£¿â£¿â ˜â¢¹â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â¡„â¢¸â ™â ¿â ¿â ¿â£¿â£¿â Ÿâ£¿â¡â ¿â£¿â£¿â¡¿â ¿â ›â¢¹â¢ â£¾â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â¢¸â£Ÿâ ˆâ “â ’â ’â ‰â ‰â €â£´â£¿â£¿â£„â €â£€â¡‰â ‰â ‰â â¢ˆâ£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â ˜â£¿â£¶â£¶â¢¶â£¿â£¿â¡·â ˜â¢¿â£¿â ¿â Ÿâ €â£¿â£¿â£¿â£·â£¶â¢»â£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â €â¢¹â£¿â£¿â£¿â£¿â¢»â¡‡â °â †â €â €â ¶â €â£¿â£¿â£¿â£¿â£¿â¢¸â£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â €â ˜â£¿â£¿â£¿â£¿â£¿â£¶â¡¶â¡·â¢¶â¢¾â¢¿â£¶â£¿â£¿â£¿â£¿â£¿â£¿â£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â €â €â¡¿â£¿â£½â£¿â¡Ÿâ¢¿â£½â£³â£Ÿâ£¾â£¯â¡¿â£‹â£¿â£Ÿâ£Ÿâ£¿â£¿â£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â €â €â €â €â¡‡â¡¿â£Žâ¢¿â¢¿â£³â£¸â ‰â ‰â â£´â£¼â¢»â£Ÿâ£¯â¡žâ¢¸â¢¸â£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â €â €â €â €â¢€â£ â¡´â šâ£¿â¡¿â£Ÿâ£¾â£¯â¡¿â â “â ’â ’â ƒâ¢¿â£ºâ£¿â£¯â£€â£¸â¢¸â£¿â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â €â €â €â €â¢€â£ â ´â ›â ‹â â €â¢¸â£¿â¢ƒâ£¿â¡¿â£¿â£¶â¡†â €â €â â¡–â ’â¢»â£¡â¡Ÿâ €â¢¹â£¸â¡Ÿâ “â¢²â ¦â£„â¡€â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â €â¢¸
â¢€â£€â£´â šâ ‰â €â €â €â €â €â£€â£¼â ›â£¿â£¿â ‡â €â ™â£¿â£Ÿâ ›â ¶â¡¿â¢¶â£¾â¡â¢¹â¡€â¡€â¢¹â¡‡â €â ¸â €â¡â ‰â ‰â “â ²â¢¦â£¤â£€â£€â¡€â €â €â €â €â €â €â €â €â¢¸
======================================================================
""")

display_status()

while human_health > 0 and terminator_health > 0:
    game_round()

# End of game and timer calculation
end_time = time.time()
survival_time = end_time - start_time

# Game over message
if human_health <= 0:
    print(f"Game Over! T-Zero has defeated you. You survived for {survival_time:.2f} seconds.")
elif terminator_health <= 0:
    print(f"Congratulations! You have defeated T-Zero. You survived for {survival_time:.2f} seconds.")
```
